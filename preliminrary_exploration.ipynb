```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=100)
clf.fit(triage[['temperature', 'heartrate', 'resprate', 'o2sat', 'sbp', 'dbp']][:int(0.8*len(triage))], triage['acuity'][:int(0.8*len(triage))])
```

```python
X_test = triage[['temperature', 'heartrate', 'resprate', 'o2sat', 'sbp', 'dbp']][int(0.8*len(triage)):]
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn import metrics

y_test = triage['acuity'][int(0.8*len(triage)):]

metrics.accuracy_score(y_test, y_pred)
```

```python
data = []

for estimator in clf.estimators_:
    data.append(estimator.predict_proba(X_test))
```

```python
# demonstrates how we can visualize the distribution of decision tree prediction vectors
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced = pca.fit_transform(data)

t = reduced.transpose()

plt.scatter(t[0], t[1])
plt.show()
```
